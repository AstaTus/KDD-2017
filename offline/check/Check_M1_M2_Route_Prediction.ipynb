```python
routes = pd.read_csv('../../data/original/routes (table 4).csv')
route_links = []
for i in range(len(routes)):
    link_elements = []

    link_seq = routes.loc[i]['link_seq']
    links = link_seq.split(',')
    for l in links:
        link_elements.append(int(l))
        
    route_links.append(link_elements)
```

```python
route_links
```

```python
def getRouteData(validation_path, route_id, date, time, index):
    links = route_links[route_id]
    period = 'morning/'
    
    if time == 1:
        period = 'afternoon/'
        
    real_middles = []
    real_alls = []
    
    predict_middles = []
    predict_tops = []
    
    for l in links:
        real_middles.append(pd.read_csv('../../data/process/link_average_arrive_time/' + period + str(l) + '.csv', index_col=0).loc[date].values)
        real_diffs.append(pd.read_csv('../../data/process/link_residual_time/' + period + str(l) + '.csv', index_col=0).loc[date].values)

        predict_middles.append(pd.read_csv(validation_path + 'prediction/afternoon/m1_' + str(l) + '.csv', index_col=0).loc[index])
        predict_tops.append(pd.read_csv(validation_path + 'prediction/afternoon/m2_' + str(l) + '.csv', index_col=0).loc[index])
        
    return real_middles, real_diffs, predict_middles, predict_tops
```

```python
#validation_path = '../../data/offline/validation_9_6/'
validation_path = '../../data/offline/validation_9_20/'
#validation_path = '../../data/offline/validation_10_11/'
```

```python
route_id = 4
date = '2016-09-26'
time=1
index = 6
```

```python
real_middles, real_diffs, predict_middles, predict_diffs = getRouteData(validation_path, route_id, date, time, index)
```

```python
fig, axes = plt.subplots(nrows=12, ncols=1, figsize=(15, 60))

t = 0
for t in range(12):
    real_link_middles = []
    real_link_diffs = []
    predict_link_middles = []
    predict_link_diffs = []
    for l in range(len(real_middles)):
        real_link_middles.append(real_middles[l][t])
        real_link_diffs.append(real_diffs[l][t])
        predict_link_middles.append(predict_middles[l][t - 6])
        predict_link_diffs.append(predict_diffs[l][t - 6] / 2)
        
    axes[t].plot(np.array(range(len(real_link_middles))), real_link_middles, 'o-', label='middle_real:' + str(int(np.sum(real_link_middles))))
    axes[t].plot(np.array(range(len(real_link_diffs))), real_link_diffs, 'o-', label='diff_real:' + str(int(np.sum(real_link_diffs))))
    if t >= 6:
        axes[t].plot(np.array(range(len(predict_link_middles))), predict_link_middles, 'o-', label='middle_predict:' + str(int(np.sum(predict_link_middles))))
        axes[t].plot(np.array(range(len(predict_link_diffs))), predict_link_diffs, 'o-', label='diff_predict:' + str(int(np.sum(predict_link_diffs))))
    axes[t].legend()

plt.show()
```
