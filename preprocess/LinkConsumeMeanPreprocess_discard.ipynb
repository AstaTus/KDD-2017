```python
import numpy as np
import pandas as pd
import datetime
```

```python
travel_link_df = pd.read_csv('../data/test/middle_normal_link.csv', parse_dates=['end_time', 'start_time', 'norm_start_time'])
```

```python
link_df = pd.read_csv('../data/original/links (table 3).csv', index_col=0)
```

```python
am_start_date = pd.to_datetime('2016-10-18')
am_end_date = pd.to_datetime('2016-10-24')

morning_link_consume_dict = {}
afternoon_link_consume_dict = {}

for link_id in link_df.index.values:
    appoint_link_df = travel_link_df[travel_link_df['link'] == link_id]
    morning_link_consumes = []
    afternoon_link_consumes = []
    for t in range(6):
        norm_date = am_start_date

        day_morning_consumes = []
        day_afternoon_consumes = []
        
        while (norm_date <= am_end_date ):
            morning_time = norm_date + datetime.timedelta(hours=6, minutes=20 * t)
            afternoon_time = norm_date + datetime.timedelta(hours=15, minutes=20 * t)

            temp_df = appoint_link_df[appoint_link_df['norm_start_time'] == morning_time]
            day_morning_consumes.extend(temp_df['consume'])

            temp_df = appoint_link_df[appoint_link_df['norm_start_time'] == afternoon_time]
            day_afternoon_consumes.extend(temp_df['consume'])

            norm_date = norm_date + datetime.timedelta(days=1)
            
        morning_link_consumes.append(np.mean(day_morning_consumes))
        afternoon_link_consumes.append(np.mean(day_afternoon_consumes))
    morning_link_consume_dict[link_id] = morning_link_consumes
    afternoon_link_consume_dict[link_id] = afternoon_link_consumes
```

```python
pd.DataFrame(morning_link_consume_dict).T.to_csv('../data/test/morning_middle_link_mean_consume.csv')
pd.DataFrame(afternoon_link_consume_dict).T.to_csv('../data/test/afternoon_middle_link_mean_consume.csv')
```
