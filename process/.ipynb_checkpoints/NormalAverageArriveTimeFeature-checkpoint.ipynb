```python
import numpy as np
import pandas as pd
import datetime
```

```python
travel_link_df = pd.read_csv('../data/preprocess/middle_normal_link.csv', parse_dates=['end_time', 'start_time', 'norm_start_time'])
```

```python
link_df = pd.read_csv('../data/original/links (table 3).csv', index_col=0)
```

# 时间区间 ：7月19日-10月24日 
#### 每个ink输出对应的两个文件 分别是上午的4个小时和下午的4个小时，计算每20分钟的平均到达时间, 测试集数据只有2小时

```python
len(link_df)
```

```python
am_start_date = pd.to_datetime('2016-07-19')
am_end_date = pd.to_datetime('2016-10-24')
link_time_dict = {}
link_count_dict = {}
for link_id in link_df.index.values:
    appoint_link_df = travel_link_df[travel_link_df['link'] == link_id]
    morning_link_time_dict = {}
    afternoon_link_time_dict = {}
    norm_date = am_start_date
    norm_time = norm_date + datetime.timedelta(hours=6)
    while (norm_date <= am_end_date ):
        morning_time = norm_date + datetime.timedelta(hours=6)
        afternoon_time = norm_date + datetime.timedelta(hours=15)
        day_morning_times = []
        day_afternoon_times = []
        for t in range(12):
            temp_df = appoint_link_df[appoint_link_df['norm_start_time'] == morning_time]
            if len(temp_df) == 0:
                day_morning_times.append(np.nan)
            else:
                day_morning_times.append(temp_df['consume'].mean())
            morning_time = morning_time + datetime.timedelta(minutes=20)
            
            temp_df = appoint_link_df[appoint_link_df['norm_start_time'] == afternoon_time]
            if len(temp_df) == 0:
                day_afternoon_times.append(np.nan)
            else:
                day_afternoon_times.append(temp_df['consume'].mean())
            afternoon_time = afternoon_time + datetime.timedelta(minutes=20)

        morning_link_time_dict[norm_date] = day_morning_times
        afternoon_link_time_dict[norm_date] = day_afternoon_times

        norm_date = norm_date + datetime.timedelta(days=1)
        
    link_time_dict[link_id] = [morning_link_time_dict, afternoon_link_time_dict]
    
```

```python
#前三天 + 后三天的均值，取到多少算多少，遇到NAN 则跳过
def getMeanTime(df, i, t, min_index, max_index):
    s = 0
    c = 0
    for k in range(1, 5):
        if (i - k) >= min_index :
            s = s + df.loc[i-k][t]
            c = c + 1
            
        if ((i + k) <= max_index) and ~np.isnan(df.loc[i + k][t]):
            s = s + df.loc[i+k][t]
            c = c + 1
    return s / c
```

```python
for link_id in link_df.index.values:
    for m in range(2):
        time_df = pd.DataFrame(link_time_dict[link_id][m]).T
        temp_df = time_df.reset_index(drop=True)
        
        for i in range(91):
            for t in range(12):
                if np.isnan(temp_df.loc[i][t]) :
                    temp_df.iloc[i][t] = getMeanTime(temp_df, i, t, 0, 90)

        for i in range(91, 98):
            for t in range(6):
                if np.isnan(temp_df.loc[i][t]) :
                    temp_df.iloc[i][t] = getMeanTime(temp_df, i, t, 0, 97)
        
        time_df[np.array(range(12))] = temp_df[np.array(range(12))].values
        if m == 0:
            time_df.to_csv('../data/feature/link_average_arrive_time/morning/' + str(link_id) + '.csv')
        else:
            time_df.to_csv('../data/feature/link_average_arrive_time/afternoon/' + str(link_id) + '.csv')
```
